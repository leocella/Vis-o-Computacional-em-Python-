Create a flowchart diagram based on a LaTeX TikZ code that shows a video processing workflow. The diagram should have 5 sequential steps connected by arrows:

1. "Vídeo de entrada (sequencia1.mp4)" - blue background
2. "Segmentação por cor (Distância Euclidiana em BGR)" - yellow background  
3. "Máscara binária + Morfologia" - green background
4. "Extração da coluna central" - orange background
5. "Contagem de objetos + Estatísticas de tempo" - red background

Each step should be in a rounded rectangle box with the specified background colors. The boxes should be connected with thick arrows flowing downward. The title should be "Fluxo de processamento da Atividade 1". Use a clean, professional style with good spacing between elements.

I'll create a flowchart diagram based on the LaTeX TikZ code showing the video processing workflow with the 5 sequential steps and specified colors.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

# Define colors matching the LaTeX specification
colors = {
    'blue': '#E6F3FF',    # blue!10
    'yellow': '#FFF9CC',  # yellow!20
    'green': '#E8F5E8',   # green!15
    'orange': '#FFE6CC',  # orange!20
    'red': '#FFE6E6'      # red!20
}

# Define box positions and text
boxes = [
    {'text': 'Vídeo de entrada\n(sequencia1.mp4)', 'color': colors['blue'], 'y': 5},
    {'text': 'Segmentação por cor\n(Distância Euclidiana em BGR)', 'color': colors['yellow'], 'y': 4},
    {'text': 'Máscara binária\n+ Morfologia', 'color': colors['green'], 'y': 3},
    {'text': 'Extração da coluna central', 'color': colors['orange'], 'y': 2},
    {'text': 'Contagem de objetos\n+ Estatísticas de tempo', 'color': colors['red'], 'y': 1}
]

# Box dimensions
box_width = 3.5
box_height = 0.6
x_center = 0

# Draw boxes
for i, box in enumerate(boxes):
    # Create rounded rectangle
    fancy_box = FancyBboxPatch(
        (x_center - box_width/2, box['y'] - box_height/2),
        box_width, box_height,
        boxstyle="round,pad=0.1",
        facecolor=box['color'],
        edgecolor='black',
        linewidth=1.5
    )
    ax.add_patch(fancy_box)
    
    # Add text
    ax.text(x_center, box['y'], box['text'], 
           ha='center', va='center', 
           fontsize=11, fontweight='normal',
           wrap=True)

# Draw arrows between boxes
arrow_props = dict(arrowstyle='->', lw=2.5, color='black')
for i in range(len(boxes) - 1):
    y_start = boxes[i]['y'] - box_height/2 - 0.05
    y_end = boxes[i+1]['y'] + box_height/2 + 0.05
    ax.annotate('', xy=(x_center, y_end), xytext=(x_center, y_start),
                arrowprops=arrow_props)

# Set axis properties
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(0.2, 5.8)
ax.set_aspect('equal')
ax.axis('off')

# Add title
plt.title('Fluxo de processamento da Atividade 1', 
          fontsize=14, fontweight='bold', pad=20)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

1. **Vídeo de entrada (sequencia1.mp4)** - Light blue background

3. **Máscara binária + Morfologia** - Light green background
4. **Extração da coluna central** - Light orange background
5. **Contagem de objetos + Estatísticas de tempo** - Light red background
